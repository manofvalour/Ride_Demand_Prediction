Circular Queue implementation

```python
class Queue:
    def __init__(self, size):
        self.__maxsize = size
        self.__que = [None]*size
        self.__front= 1
        self.__rear = 0
        self.__nitems = 0

    def insert(self, value):
        if self.__nitems >= self.__maxsize:
            print('Queue overflow')

        self.__rear+=1
        if self.__rear == self.__maxsize:
            self.__rear = 0
        self.__que[self.__rear]= value
        self.__nitems+=1

        print('Item added successfully!')

    def remove(self):
        if self.__nitems == 0:
            raise Exception('Queue is empty')

        front = self.__que[self.__front]  ## viewing the value to remove
        self.__que[self.__front]= None   ## removing the first item
        self.__front+=1     ## shifting the first index forward

        if self.__front ==self.__maxsize:
            self.__front = 0
        self.__nitems-=1

        print('Item successfully removed')
        return 

    def peek(self):
        if self.__nitems==0:
            return None
        else:
            return self.__que[self.__front]

    def isEmpty(self): 
        return self.__nitems == 0

    def isFull(self): 
        return self.__nitems == self.__maxsize
    
    def __len__(self): 
        return self.__nitems
    
    def __str__(self):
        ans = "["
        for i in range(self.__nitems):
            if len(ans) > 1:
                ans += ", "
            j = i + self.__front
            if j >= self.__maxsize:
                j -= self.__maxsize
                ans += str(self.__que[j])
            ans += "]"
        return ans      
        
```
